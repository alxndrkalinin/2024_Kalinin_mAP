```python
%load_ext autoreload
%autoreload 2

import numpy as np

from map_utils.plot import set_plotting_style, plot_map_x3
from orf_utils import load_orf_map_results
```

```python
set_plotting_style()
```

```python
map_results = load_orf_map_results()
activity_map = map_results.query(
    "Modality == 'ORF' and retrieval == 'Replicates'"
).copy()
consistency_map = map_results.query(
    "Modality == 'ORF' and retrieval == 'CORUM Complex'"
).copy()
```

```python
activity_map["-log10(mAP p-value)"] = -np.log10(activity_map["corrected_p_value"])
activity_map.rename(
    {
        "mean_average_precision": "mAP",
        "below_corrected_p": "p < 0.05",
    },
    axis=1,
    inplace=True,
)
activity_map["const"] = "dataset: cpg0016[orf]"
plot_map_x3(activity_map, "const", "", pr_x=0.45, pr_y=0.02, l_x=1.1, l_y=0.58)
```

```python
consistency_map["-log10(mAP p-value)"] = -np.log10(consistency_map["corrected_p_value"])
consistency_map.rename(
    {
        "mean_average_precision": "mAP",
        "below_corrected_p": "p < 0.05",
    },
    axis=1,
    inplace=True,
)
consistency_map["const"] = "dataset: cpg0016[orf]"
plot_map_x3(consistency_map, "const", "", pr_x=0.45, pr_y=0.02, l_x=1.1, l_y=0.58)
```
